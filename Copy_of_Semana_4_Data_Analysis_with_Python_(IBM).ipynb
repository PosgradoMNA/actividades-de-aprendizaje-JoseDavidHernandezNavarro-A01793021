**Semana 4: Data Analysis with Python (IBM)**

```python
import pandas as pd #Importamos nuestras librerias "pandas" y "numpy" y les asignamos una abreviacion
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv" #Marcamos la direccion de nuestro dataset
df = pd.read_csv(path, header=None) #leemos el dataset
```

```python
print("Las primeras cinco filas del dataframe son") #Imprime texto
df.head(5) #Con este comando imprimimos las priemras cinco filas del dataframe
```

```python
print("Las ultimas 10 filas del dataframe son:\n") 
df.tail(10) #Coneste codigo imprimimos las ultimas 10 columnas (el valor numerico se sustituye por el numero de filas a imprimier)
```

```python
#Creamos una lista en la cual meteremos los encabezados a acoplar en el dataframe
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.head(10) #Reemplazamos los encabezados por los incluidos en la lista "headers" e imprimimos las primeras 10 filas
```

```python
df1=df.replace('?',np.NaN) #Reemplazamos el simolo "?" por la expresion NaN a modo de acerlo mas integrado al manejo de datos
```

```python
df=df1.dropna(subset=["price"], axis=0) #Podemos eliminar solo los valores faltantes de cualquier columna, en este caso es a precio
df.head(20)
```

```python
print(df.columns) #Si el dataframe ya incluye los encabezados podemos saber el nombre de cada columna de esta forma
```

```python
print(df.dtypes) #Usamos esta funcion para saber el tipo de dato que maneja cada columna
```

```python
df.describe() #Con este comando obtenemos valores interesantes de nuestro dataframe 
```

```python
df.describe(include = "all") #con esta funcion podenmos teber una descripcopn general de nuestro dataframe
```

```python
df[['length', 'compression-ratio']].describe() #Podemos obtener nuevamente los valores especificos pero ahora de las columnas que deseemos mencionadolas por su encabezado
```
